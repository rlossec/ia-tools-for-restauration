```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from wordcloud import WordCloud
import pickle
```

```python
DATA_PATH = '../data/2_preprocessed_data/'
```

## Chargement des donn√©es

```python
with open(DATA_PATH + "corpus_bad_tokens", "rb") as fp:
    corpus_bad_tokens_2 = pickle.load(fp)
```

```python
with open(DATA_PATH + "corpus_good_tokens", "rb") as fp:
    corpus_good_tokens_2 = pickle.load(fp)
```

```python
with open(DATA_PATH + "corpus_sample_tokens", "rb") as fp:
    corpus_sample_tokens_2 = pickle.load(fp)
```

```python
with open(DATA_PATH + "corpus_bad_without_context", "rb") as fp:
    corpus_bad_tokens_3 = pickle.load(fp)
```

```python
with open(DATA_PATH + "corpus_good_without_context", "rb") as fp:
    corpus_good_tokens_3 = pickle.load(fp)
```

```python
with open(DATA_PATH + "corpus_sample_without_context", "rb") as fp:
    corpus_sample_tokens_3 = pickle.load(fp)
```

## Premier WordCloud

```python
wordcloud = WordCloud(background_color='white',
                      stopwords=[],
                      max_words=50).generate(" ".join(corpus_bad_tokens_2))
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
```

```python
wordcloud = WordCloud(background_color='white',
                      stopwords=[],
                      max_words=50).generate(" ".join(corpus_good_tokens_2))
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
```

```python
wordcloud = WordCloud(background_color='white',
                      stopwords=[],
                      max_words=50).generate(" ".join(corpus_sample_tokens_2))
plt.imshow(wordcloud)
plt.axis("off")
plt.show();
```

On voit que beaucoup de mots sont communs aux diff√©rents nuages de mots : order, food, one, ...  
Construisons des mots de contexte pour que ces mots soient filtr√©s.
